```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
n=1000
x = np.linspace(0,10,n)
y = 0.5*x+2
noise = np.random.normal(0,0.1,n)
y = y + noise
plt.plot(x,y)
plt.legend(["Dados Reais"])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = x.reshape(-1,1)
y = y.reshape(-1,1)
model = LinearRegression()
model.fit(x,y)
yhat = model.predict(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,yhat)
plt.legend(["Dados Reais",'Previso'])
plt.show()
```

```python
import joblib
joblib.dump(model,'../server/model.joblib')
```

```python
model2 = joblib.load('../server/model.joblib')
```

```python
yhat = model2.predict(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,yhat)
plt.legend(["Dados Reais",'Previso'])
plt.show()
```
